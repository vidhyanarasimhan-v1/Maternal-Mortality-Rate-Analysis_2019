```python
!pip install wbdata
import wbdata
import pandas as pd
import numpy as np
from datetime import datetime
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
indicators={"SH.STA.MMRT":"Maternal_Mortality",
            "NY.GDP.PCAP.CD":"GDP_Per_Capita",
            "SH.XPD.CHEX.GD.ZS":"Current_Health_Expenditure",
            "SH.ANM.ALLW.ZS":"Anemia",
            "SE.ADT.LITR.ZS":"Adult_Literacy_Rate"}
data=wbdata.get_dataframe(indicators,country="all",date='2019')
data=data.dropna()
data=data[data['Maternal_Mortality']>0]
print(data)
data['log_MMR']=np.log(data['Maternal_Mortality'])
data['log_gdp']=np.log(data['GDP_Per_Capita'])
Y=data['log_MMR']
X=data[['log_gdp','Current_Health_Expenditure','Anemia','Adult_Literacy_Rate']]
X=sm.add_constant(X)
vif=pd.DataFrame()
vif["Variable"]=X.columns
vif["VIF"]=[variance_inflation_factor(X.values,i)for i in range(X.shape[1])]
print(vif)
model=sm.OLS(Y,X).fit()
print(model.summary())
plt.scatter(model.fittedvalues,model.resid)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()
sns.histplot(model.resid,kde=True)
```
